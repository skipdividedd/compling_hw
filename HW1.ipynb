```python
import re
```

```python
def remove_tags(text):
    pattern1 = '(?:<h\d*?>)' # pattern for head tags
    result1 = re.sub(pattern1, '\n', text) # make new line for head tags
    pattern2 = '(?:</?\w+.*?>)' # pattern for all other tags and meta info
    result2 = re.sub(pattern2, '', result1) # delete tags and meta info
    return result2
```

## 4. Pymorphy vs Mystem (2 балла)

Лемматизируйте очищенный текст с помощью mystem и pymorphy. Сравните результаты: есть ли разница, что лучше. 

Для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат.

(не нашла токенизацию в Mystem, использовала токенизацию пайморфи).

```python
from pymystem3 import Mystem
import pymorphy2
```

```python
from pymorphy2 import MorphAnalyzer
import pymorphy2.tokenizers
from pymorphy2.tokenizers import simple_word_tokenize
from pymorphy2.shapes import is_punctuation
```

```python
def mystem_lemmatize(mystem, text):  #plain mystem
    lemmas = mystem.lemmatize(text)
    return lemmas

def mystem_lemmatize_tokenized(mystem, text): #mystem with pymorphy tokenization
    lemmas = [mystem.lemmatize(token)[0] for token in simple_word_tokenize(text) if not is_punctuation(token)]
    return lemmas

def pymorphy_lemmatize(morph, text): #pymorhpy with with pymorphy tokenization
    lemmas = []
    for token in simple_word_tokenize(text):
        if not is_punctuation(token):
            lemmas.append(morph.parse(token)[0].normal_form)   
    return lemmas
```

```python
mystem = Mystem() #mystem
morph = MorphAnalyzer() #pymorphy
```

```python
%time
lemmas_mystem = mystem_lemmatize(mystem, new_text)
```

```python
%time
lemmas_mystem_tokenized = mystem_lemmatize_tokenized(mystem, new_text)
```

```python
%time
lemmas_pymorphy = pymorphy_lemmatize(morph, new_text)
```

```python
print(''.join(lemmas_mystem)) #plain mystem
```

```python
print(*lemmas_mystem_tokenized) #mystem with pymorhpy tokenization
```

```python
print(*lemmas_pymorphy)
```

Кажется, что plain mystem (с собственной токенизацией) работает лучше всего, поскольку по умолчанию стоит disambiguation = True. 

Тем не менее, чтобы сравнить саму работу Mystem и Pymorpy можно сравнить результаты лемматизации, получившиеся при использовании токенайзера из пайморфи. 

Например, фраза "выводить эти данные в различных формах" -- 

mystem: "выводить этот данные в различный форма", 

pymorphy: "выводить этот дать в различный форма". 

Кроме того, pymprhy неправильно лемматизировал слово "деньги" -- получилось "деньга". 

Тем не менее, есть случаи, когда pymorphy дает лучший результат: например, "теплее" -- 
"тепло" (mystem), 
"теплый" (pymorphy).
